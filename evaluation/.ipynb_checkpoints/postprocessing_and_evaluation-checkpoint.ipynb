```python
import os
import cv2
import glob
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

from scipy.linalg import sqrtm
from skimage.transform import resize
from tensorflow.keras.applications.inception_v3 import InceptionV3, preprocess_input
from tensorflow.keras.models import Model

def preprocess_image(image):
    # Resize image to (299, 299) for Inception-v3
    image_resized = resize(image, (299, 299))
    # Preprocess the image for Inception-v3
    image_preprocessed = preprocess_input(image_resized)
    return image_preprocessed

def apply_mask(image, mask):
    # Apply mask to image
    masked_image = image.copy()
    masked_image[~mask] = 0  # Set non-masked areas to 0
    return masked_image

def extract_features(image, model):
    # Preprocess images
    preprocessed_image = np.array(preprocess_image(image))
    preprocessed_image = np.array([preprocessed_image])
    #preprocessed_image = np.expand_dims(preprocessed_image, axis=-1)
    print(preprocessed_image.shape)
    # Extract features
    features = model.predict(preprocessed_image)
    return features
```

```python
def calculate_fid(real_images, generated_images):
    # Load InceptionV3 model pre-trained on ImageNet
    inception_model = InceptionV3(include_top=False, pooling='avg', input_shape=(299, 299, 3))

    # Preprocess images and get their feature representations
    real_features = inception_model.predict(preprocess_input(real_images.astype('float32')))
    gen_features = inception_model.predict(preprocess_input(generated_images.astype('float32')))

    # Calculate mean and covariance of features
    mu1, sigma1 = np.mean(real_features, axis=0), np.cov(real_features, rowvar=False)
    mu2, sigma2 = np.mean(gen_features, axis=0), np.cov(gen_features, rowvar=False)

    # Calculate FID score
    diff = mu1 - mu2
    covmean, _ = sqrtm(sigma1 @ sigma2, disp=False)
    if np.iscomplexobj(covmean):
        covmean = covmean.real

    fid_score = np.sum(diff**2) + np.trace(sigma1 + sigma2 - 2 * covmean)
    return fid_score

# Example usage
real_images = np.random.rand(100, 299, 299, 3)  # Replace with your real images
generated_images = np.random.rand(100, 299, 299, 3)  # Replace with your generated images

fid_score = calculate_fid(real_images, generated_images)
print("FID Score:", fid_score)
```

```python
def calculate_fid_single_image(true_image, fake_image):
    # Convert images to PyTorch tensors and normalize
    true_tensor = to_tensor(true_image).float()
    fake_tensor = to_tensor(fake_image).float()

    # Flatten the images
    true_image_flat = true_tensor.view(-1)
    fake_image_flat = fake_tensor.view(-1)

    # Compute mean vectors for real and generated images
    mu1 = torch.mean(true_image_flat, axis=0)
    mu2 = torch.mean(fake_image_flat, axis=0)

    # Compute covariance matrices for real and generated images
    sigma1 = torch.cov(true_image_flat, rowvar=False)
    sigma2 = torch.cov(fake_image_flat, rowvar=False)

    # Compute squared Mahalanobis distance
    diff = mu1 - mu2
    cov_mean_sqrt = sqrtm(sigma1 @ sigma2)

    # Calculate FID score
    fid_score = np.dot(diff, diff) + np.trace(sigma1 + sigma2 - 2 * cov_mean_sqrt)
    return fid_score
```
