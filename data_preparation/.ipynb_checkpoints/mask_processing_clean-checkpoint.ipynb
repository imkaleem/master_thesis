```python
import cv2

def process_image_raw(filename):
    # Read the original RGB image
    vessel_image = cv2.imread(filename)

    # Convert to grayscale
    gray_image = cv2.cvtColor(vessel_image, cv2.COLOR_BGR2GRAY)

    # Apply a binary threshold
    _, binary_image = cv2.threshold(gray_image, 125, 255, cv2.THRESH_BINARY)

    # Apply pyramid up (upsample the image)
    blurred_image = cv2.pyrUp(binary_image)

    # Apply median blur
    for _ in range(15):
        blurred_image = cv2.medianBlur(blurred_image, 7)

    # Apply pyramid down (downsample the image)
    blurred_image = cv2.pyrDown(blurred_image)

    # Apply a binary threshold
    _, final_image = cv2.threshold(blurred_image, 200, 255, cv2.THRESH_BINARY)

    # Erode the binary mask to reduce the size of the white regions
    kernel = np.ones((4, 4), np.uint8)  # You can adjust the kernel size for more or less erosion
    eroded_image = cv2.erode(final_image, kernel, iterations=1)

    # Convert back to RGB
    final_image_rgb = cv2.cvtColor(eroded_image, cv2.COLOR_GRAY2BGR)

    return final_image_rgb

def process_image(image):
    # Read the original image
    #vessel_image = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
    vessel_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply a binary threshold
    _, vessel_image = cv2.threshold(vessel_image, 125, 255, cv2.THRESH_BINARY)

    # Apply pyramid up (upsample the image)
    blurred_image = cv2.pyrUp(vessel_image)

    # Apply median blur 15 times
    for _ in range(15):
        blurred_image = cv2.medianBlur(blurred_image, 7)

    # Apply pyramid down (downsample the image)
    blurred_image = cv2.pyrDown(blurred_image)

    # Apply a binary threshold
    _, blurred_image = cv2.threshold(blurred_image, 200, 255, cv2.THRESH_BINARY)
    print(blurred_image.shape)

    return blurred_image




# Replace with the path to your image
#processed_image = process_image(image_paths[3])
#processed_image.dtype, plt.imshow(processed_image)
```

```python
def generate_wb_mask(image_path, itr = 3):
    mask_3c = generate_mask3c(image_path, itr)
    processed_image = process_image(mask_3c)
    
    mask_3c = convert_binary_to_3c_mask(~processed_image)       #new_mask
    mask_3c = mask_3c.astype("uint8")
    
    cv2.imwrite(image_path.replace(image_path[-4], "_mask."), mask_3c)
```

```python
def process_image(filename):
    # Read the original RGB image
    vessel_image = cv2.imread(filename)

    # Convert to grayscale
    gray_image = cv2.cvtColor(vessel_image, cv2.COLOR_BGR2GRAY)

    # Apply a binary threshold
    _, binary_image = cv2.threshold(gray_image, 125, 255, cv2.THRESH_BINARY)

    # Apply pyramid up (upsample the image)
    blurred_image = cv2.pyrUp(binary_image)

    # Apply Gaussian blur
    for _ in range(15):
        blurred_image = cv2.GaussianBlur(blurred_image, (7, 7), 0)

    # Apply pyramid down (downsample the image)
    blurred_image = cv2.pyrDown(blurred_image)

    # Apply a binary threshold
    _, final_image = cv2.threshold(blurred_image, 200, 255, cv2.THRESH_BINARY)

    # Convert back to RGB
    final_image_rgb = cv2.cvtColor(final_image, cv2.COLOR_GRAY2BGR)

    return final_image_rgb

# Example usage
processed_image_raw = process_image(image_paths[1])
plt.imshow(processed_image_raw)
```

```python
def resize_and_save_image(image_path, target_size=(960, 960)):
    # Load the image
    image = cv2.imread(image_path)

    # Resize the image
    resized_image = cv2.resize(image, target_size)

    # Save the resized image
    output_path = image_path.replace(image_path[-4], "_new.")
    cv2.imwrite(output_path, resized_image)
    print(f"Resized image saved to: {output_path}")
```
