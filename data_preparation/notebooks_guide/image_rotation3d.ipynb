```python
def create_alpha_channel(image_path, mask):
    # Read the image with a transparent background (PNG or GIF)
    image = cv2.imread(image_path, cv2.IMREAD_UNCHANGED)

    # If the image already has an alpha channel, return it
    if image.shape[2] == 4:
        return image

    # Create an alpha channel based on a mask
    #mask = np.all(image[:, :, :3] == [255, 255, 255], axis=2)  # Assuming white is the background color
    #mask = masks[0].masks.data[0]
    plt.imshow(mask)
    alpha_channel = np.where(mask, 0, 255).astype(np.uint8)

    # Add the alpha channel to the original image
    image_with_alpha = np.dstack([image, alpha_channel])

    # Crop the image based on the alpha channel
    alpha_channel_binary = (alpha_channel != 0).astype(np.uint8)  # Ensure it's uint8
    contours, _ = cv2.findContours(alpha_channel_binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # Find the bounding box of the contours
    bounding_box = cv2.boundingRect(contours[0])

    # Extract the region of interest (ROI) based on the bounding box
    cropped_image_with_alpha = image_with_alpha[bounding_box[1]:bounding_box[1]+bounding_box[3],
                                                 bounding_box[0]:bounding_box[0]+bounding_box[2]]
    return image_with_alpha


# def create_alpha_channel(image_path):
#     image = cv2.imread(image_path, cv2.IMREAD_UNCHANGED)

#     if image.shape[2] == 4:
#         return image

#     # Convert the image to grayscale
#     gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#     # Create a mask based on a threshold
#     _, mask = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY)

#     # Dilate the mask to capture anti-aliased edges
#     kernel = np.ones((3, 3), np.uint8)
#     mask = cv2.dilate(mask, kernel, iterations=1)

#     # Create an alpha channel based on the dilated mask
#     alpha_channel = mask.astype(np.uint8)

#     # Add the alpha channel to the original image
#     image_with_alpha = np.dstack([image, alpha_channel])

#     return image_with_alpha

def paste_image(background, foreground, position=(0, 0)):
    """
    Paste a smaller image (foreground) onto a larger image (background) at the specified position.
    
    Parameters:
        - background (numpy.ndarray): The larger image.
        - foreground (numpy.ndarray): The smaller image to be pasted.
        - position (tuple): The position (x, y) where the top-left corner of the smaller image will be placed.
        
    Returns:
        - numpy.ndarray: The resulting image with the smaller image pasted onto the larger one.
    """
    h, w = foreground.shape[:2]
    background = cv2.resize(background, (1024, 1024))
    
    center_x = np.random.randint(0, 1024)
    center_y = np.random.randint(0, 1024)
    
    # Ensure the position is within the bounds of the background image
    y, x = (center_y, center_x)
    y = max(0, min(y, background.shape[0] - h))
    x = max(0, min(x, background.shape[1] - w))
    
    # Create a region of interest (ROI) for the smaller image on the background
    roi = background[y:y + h, x:x + w]
    
    # Create a mask for the foreground image
    mask = (foreground[:, :, 3] / 255.0).reshape((h, w, 1))
    
    # Blend the images using the alpha channel of the foreground
    blended = roi * (1 - mask) + foreground[:, :, :3] * mask
    
    # Update the background with the blended region
    background[y:y + h, x:x + w] = blended
    
    return background
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def perspective_transform_auto(image_path):
    # Read the input image
    image = cv2.imread(image_path)
    image = invert_image(image)

    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Detect features in the image (e.g., corners)
    corners = cv2.goodFeaturesToTrack(gray, maxCorners=4, qualityLevel=0.01, minDistance=50)

    # Convert the corners to floating point format
    corners = np.float32(corners)

    # Define the destination points for the perspective transformation
    width, height = image.shape[1], image.shape[0]
    dst_points = np.float32([[0, 0], [width, 0], [width, height], [0, height]])

    # Compute the perspective transformation matrix
    M, _ = cv2.findHomography(corners, dst_points)

    # Apply the perspective transformation to the image
    warped_image = cv2.warpPerspective(image, M, (width, height))

    # Display the original and warped images using plt.imshow()
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(warped_image, cv2.COLOR_BGR2RGB))
    plt.title('Warped Image')
    plt.axis('off')

    plt.show()

# Test the perspective_transform_auto function with an example image
perspective_transform_auto(obj_img_path)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def find_object_corners(image):
    # Convert the image to grayscale
    contours = find_contours(image)

    # Find the contour with the largest area
    max_contour = max(contours, key=cv2.contourArea)
    
    # Find the bounding rectangle of the largest contour
    x, y, w, h = cv2.boundingRect(max_contour)
    
    # Draw the bounding rectangle
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
    plt.imshow(image)

    # Get the four corners of the bounding rectangle
    corners = np.array([[x, y], [x + w, y], [x + w, y + h], [x, y + h]])

    return corners

def perspective_transform_auto(image_path, coverage_threshold=0.4):
    # Read the input image
    image = cv2.imread(image_path)
    image = invert_image(image)

    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Detect features in the image (e.g., corners)
    corners = find_object_corners(image)

    # Convert the corners to floating point format
    corners = np.float32(corners)
    print(corners)

    # Define the destination points for the perspective transformation
    width, height = image.shape[1], image.shape[0]
    
    # Adjust destination points for a slight perspective transformation
    dst_points = np.float32([[0, 0],
                             [width, 0],
                             [width, height],
                             [0, height]])

    # Compute the perspective transformation matrix
    M = cv2.getPerspectiveTransform(corners, dst_points)

    # Apply the perspective transformation to the image
    warped_image = cv2.warpPerspective(image, M, (width, height))
    print(warped_image.shape)

    # Calculate the coverage percentage of the object in the warped image
    object_coverage = np.count_nonzero(cv2.cvtColor(warped_image, cv2.COLOR_BGR2GRAY)) / (width * height)

    # Display the original and warped images if coverage meets the threshold
    if object_coverage >= coverage_threshold:
        plt.figure(figsize=(10, 5))
        plt.subplot(1, 2, 1)
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        plt.title('Original Image')
        plt.axis('off')

        plt.subplot(1, 2, 2)
        plt.imshow(cv2.cvtColor(warped_image, cv2.COLOR_BGR2RGB))
        plt.title('Warped Image')
        plt.axis('off')

        plt.show()
    else:
        print("Object coverage in the warped image is below the threshold.")

# Test the perspective_transform_auto function with an example image
perspective_transform_auto(obj_img_path)
```

```python
import cv2
import numpy as np
import os

def project_image_to_3d(image_path):
    # Load the input image
    image = cv2.imread(image_path)
    image = invert_image(image)

    # Define the dimensions of the 3D plane
    plane_width = 500
    plane_height = 500

    # Resize the input image to fit the dimensions of the 3D plane
    resized_image = cv2.resize(image, (plane_width, plane_height))

    # Project the resized image onto a 3D plane
    plane = np.zeros((plane_height, plane_width, 3), dtype=np.uint8)
    plane[:resized_image.shape[0], :resized_image.shape[1]] = resized_image

    return plane

def rotate_image_3d(image, angle_x, angle_y, angle_z):
    # Define rotation matrices for each axis
    rotation_matrix_x = np.array([[1, 0, 0],
                                   [0, np.cos(angle_x), -np.sin(angle_x)],
                                   [0, np.sin(angle_x), np.cos(angle_x)]])

    rotation_matrix_y = np.array([[np.cos(angle_y), 0, np.sin(angle_y)],
                                   [0, 1, 0],
                                   [-np.sin(angle_y), 0, np.cos(angle_y)]])

    rotation_matrix_z = np.array([[np.cos(angle_z), -np.sin(angle_z), 0],
                                   [np.sin(angle_z), np.cos(angle_z), 0],
                                   [0, 0, 1]])

    # Rotate the image around each axis
    rotated_image = cv2.warpPerspective(image, rotation_matrix_x, (image.shape[1], image.shape[0]))
    rotated_image = cv2.warpPerspective(rotated_image, rotation_matrix_y, (image.shape[1], image.shape[0]))
    rotated_image = cv2.warpPerspective(rotated_image, rotation_matrix_z, (image.shape[1], image.shape[0]))

    return rotated_image

def rotate_and_save_images_3d(input_image_path, output_folder, angle_increment=10):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Load the input image and project it onto a 3D plane
    image_3d = project_image_to_3d(input_image_path)

    # Define angles for rotation around each axis
    angles = np.arange(0, 361, angle_increment)

    # Rotate the image around each axis and save the rotated images
    saved_image_paths = []
    for angle_x in angles:
        for angle_y in angles:
            # Rotate the image
            if angle_x == angle_y:
                    
                rotated_image = rotate_image_3d(image_3d, np.radians(angle_x), np.radians(angle_y), np.radians(0))
    
                # Save the rotated image
                output_path = os.path.join(output_folder, f'rotated_x{angle_x}_y{angle_y}_z{0}.jpg')
                cv2.imwrite(output_path, rotated_image)
                saved_image_paths.append(output_path)

    return saved_image_paths


saved_paths = rotate_and_save_images_3d(obj_img_path, output_folder)
print("Saved images:", saved_paths)
```
